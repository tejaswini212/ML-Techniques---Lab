```python
# Importing necessary libraries
import numpy as np
import pandas as pd
```

```python
# Lets get the dataset inside a dataframe
data = pd.read_csv('Data/crx.data',header = None)
```

```python
# Lets define the column names
col_names = ['col' + str(s) for s in range(1,17)]
```

```python
# And rename the columns as desired
data.columns = col_names
```

```python
# We can take a look at last 10 rows of our data
data.tail(10)
```

```python
# replace '?' with nan -- not a number
data = data.replace('?',np.nan)
```

```python
# retrieve info on data ---data types of features in the data
data.info()
```

Our dataset has features of type object(string), integer and float(continuous)

```python
# Making the data more understandable with the help of mapping
data['col16'] = data['col16'].map({'+':'P','-':'N'})
```

```python
# And we can see that last column's values are changed accordingly
data.head()
```

```python
# Lets get the categorical/object-type variables
cat_columns = [c for c in data.columns if data[c].dtypes == 'O'] # Here 'O' referes to features with data type as 'Object'
```

```python
# We'll have a look at what our data in cat_columns actually is
data[cat_columns].head()
```

Looks like column 2 and column 14 are incorrectly assigned object and we will recast them as float now

```python
#recast some variables to their correct types
data['col2'] = data['col2'].astype('float')
data['col14'] = data['col14'].astype('float')
```

```python
# Lets get the categorical/object-type variables
cat_columns = [c for c in data.columns if data[c].dtypes == 'O']
```

```python
data[cat_columns].head()
```

```python
# And we can also take a look at columns that are numerical
numerical = [i for i in data.columns if i not in cat_columns]
print(numerical)
```

```python
data[numerical].head()
```
